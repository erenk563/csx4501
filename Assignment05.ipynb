```python
#https://www.kaggle.com/imdevskp/corona-virus-report
import pandas as pd

df=pd.read_csv("country_wise_latest.csv")
col1=pd.DataFrame(df["Confirmed"])
```

```python
#Mean
col1.mean()
#Median
col1.median()
#Mode
col1.mode()



#calculate 10% trimmed mean
from scipy import stats
stats.trim_mean(col1, 0.1)
```

```python
#Plot

col1.plot.hist(bins=100)

# İf we look at the median and the mean they are so far from each other it means that data points are not distrubuted well.
#İn other words it's too far to be normal distrubution
```

```python
#MAD
col1.mad()
#STD
col1.std()

#IQR
Q1 = col1.quantile(0.25)
Q3 = col1.quantile(0.75)
IQR = Q3 - Q1
IQR

#Box plot
col1.boxplot(fontsize=10)

#box plot and histogram consistent,because my data is negatively skeewed,and both plots are confirm this fact.
```

```python
#My Data set Doesnt have categorical value,to proceed ı am gonna make one "from Deaths / 100 Cases"

df.loc[df['Deaths / 100 Cases'] >df["Deaths / 100 Cases"].mean(), 'Categorical_Value'] = 1
df.loc[df['Deaths / 100 Cases'] <df["Deaths / 100 Cases"].mean(), 'Categorical_Value'] = 0
```

```python
col2=df['Categorical_Value']
#mean
col2.mean()

#quantiles
Q1 = col2.quantile(0.25)
Q3 = col2.quantile(0.75)
# Bar plot
col2.plot.bar()

#plot with col 1 and 2

#Scatter plot
df.plot.scatter(x="Confirmed",y="Categorical_Value")
```
