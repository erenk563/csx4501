```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
df=load_diabetes(return_X_y=True,as_frame=True)
x=df[0]
y=df[1]
df1=pd.concat([x,y],axis=1)
df1
```

```python
#Summary
df1.describe()
```

```python
#Scatter
df1.plot.scatter(x="age",y="target")
```

```python
#histogram
df1["s1"].plot.hist()
```

```python
##### top3 correlation
df1.corr()["target"].nlargest(n=4)[1:4]
x=df1[["bmi","s5"]]
```

```python
#Test Train
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

# LR
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

lr.fit(x_train,y_train)
predict=lr.predict(x_test)
```

```python
#Linear plot
import seaborn as sns

sns.regplot(x=lr.predict(x_test), y=y_test);
```

```python
from sklearn.metrics import mean_squared_error

mean_squared=mean_squared_error(y_test,predict)

np.sqrt(mean_squared)
```
