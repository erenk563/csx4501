```python
#https://www.kaggle.com/imdevskp/corona-virus-report

import pandas as pd
df=pd.read_csv("country_wise_latest.csv")
df
```

```python
#Calculating every numeric columns mean

df.mean()
```

```python
# searching with a condition
df[df["Confirmed"]>5000]
```

```python
#Checking general info of data set

df.info()
```

```python
#summary statistics

df.describe()
```

```python
#Calculating statistics by groups

df.groupby("WHO Region").mean()
```

```python
#Finding a values which  is above average in confirmed cases
for i in range(0,df.shape[0]):
    if df["Confirmed"][i]>df["Confirmed"].mean():
        print(df["Confirmed"][i])
```

```python
#Writing and applying function with apply method(normalization)
df["Confirmed"].apply(lambda x:x/df["Confirmed"].max())
```
